```python
import pickle
import numpy as np
from sklearn.linear_model import LinearRegression

with open("capacitor_dataset.pickle", "rb") as f:
    train_data = pickle.load(f)
```

```python
t = train_data[:,:-1]
Q = train_data[:,-1]
ln_Q = np.log(Q)

model = LinearRegression()
model.fit(t, ln_Q)

m = model.coef_[0]
b = model.intercept_

C = np.exp(b)
C /= 5

R = -1/C
R /= m

print(C)
print(R)
```
